## The three ingredients

Data science sits at the intersection of three disciplines. Remove any one of them and you no longer have data science — you have something else.

The three ingredients are **statistics**, **programming**, and **domain knowledge**.

Statistics gives you the tools to draw conclusions from data without being fooled by randomness. Programming gives you the ability to work with data at scale — to clean it, transform it, and run analyses that would take years to do by hand. Domain knowledge is what tells you which questions are worth asking in the first place, and whether the answers you get make sense.

## Data science vs data analytics vs machine learning vs AI

These four terms get used interchangeably, but they mean different things. Here is a plain English breakdown.

**Data analytics** focuses on describing what has already happened. It answers questions like: "How many customers churned last quarter? Which products sold best in the Northeast?" The output is usually a report or dashboard. The tools are SQL, Excel, and visualization platforms like Tableau.

**Data science** goes further. It uses statistical methods and machine learning to predict what will happen, find patterns that are not obvious, and recommend what to do. A data scientist might ask: "Given this customer's behavior, what is the probability they will churn in the next 30 days?"

**Machine learning** is a technique used within data science. It refers specifically to algorithms that learn patterns from data without being explicitly programmed with rules. Every machine learning project is a data science project — but not every data science project uses machine learning.

**Artificial intelligence** is the broadest term. It describes any system that performs tasks that typically require human intelligence — recognizing faces, understanding language, playing games. Machine learning is the dominant approach to building AI systems today, but AI is the goal, and ML is the method.

## What data scientists actually do

The glamorized version of a data scientist's day looks like this: wake up, train a neural network, achieve breakthrough results, collect prize.

The real version looks like this:

Most of the morning is spent figuring out why a column that should contain ages has values like "N/A", "unknown", "-1", and "999". A significant portion of the afternoon is spent in a meeting explaining why the model you built last month no longer performs as well — because the data changed, not because the model is broken.

Here is what data scientists actually spend their time on:

**Data wrangling.** Joining tables from different systems. Fixing inconsistent formatting. Handling missing values. Deciding what to do with outliers. This is not glamorous, but it is the work.

**Exploratory data analysis.** Looking at the data from multiple angles before deciding what to model. Plotting distributions. Calculating summary statistics. Asking "does this make sense?" over and over.

**Communicating findings.** Writing up results, presenting charts, explaining what a model does and does not tell you, convincing stakeholders to act on what you found.

**Modeling.** Yes, this happens too — but it is usually a smaller fraction of the total time than people expect.

> **The scientist in "data scientist" is real — it means forming hypotheses, testing them, and being honest about what the evidence says.**

## The honest truth about data science

Here is something that every data science course should tell you on day one, and most do not.

**Roughly 80% of a data scientist's work is data cleaning and preparation. Maybe more.**

Raw data is almost never ready to model. It comes from multiple sources with different formats. It has missing values, duplicates, inconsistencies, and errors. It has columns that mean one thing in one system and something different in another. Getting data into a shape where you can actually learn from it is hard, painstaking, careful work.

This is not bad news. Here is how to think about it: the teams that are good at data preparation are the ones whose models actually work in production. The teams that rush past it spend months debugging model failures that were caused by dirty data all along.

Every hour you spend learning how to clean and understand data is an hour that will pay off repeatedly throughout your career.

> **Data science is not about algorithms — it's about asking better questions with data.**
